```python
data=data[['state','family_history','treatment']]
data=data.groupby(['state','family_history','treatment']).size().reset_index().rename(columns={0:'count'})
data.head(50)
```

```python
fam_treatment=data[(data['treatment']==1.0) & (data['family_history']==1.0)].reset_index(drop=True)
fam_treatment=fam_treatment.sort_values(by='count',ascending=False)
fam_treatment
```

```python
df=pd.read_csv('rate',index_col=[0])
df
```

```python
df=df.sort_values(by='rate',ascending=False)
df.reset_index(inplace=True)
```

```python
fig1, ax = plt.subplots(figsize=(20,10))

# Example data
people = df['state']
y_pos=range(0,43)

ax.barh(y_pos,df['rate'], align='center',color='dodgerblue')
ax.set_yticks(y_pos)
ax.set_yticklabels(people)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Number of people')
ax.set_title('Number of people that did treatment and have family history')

for i, v in enumerate(df['rate']):
    ax.text(v+0.1, i + .25, str(v), color='darkorange', fontweight='bold')

#plt.savefig('patientsandfamily.png')
    
plt.show()
```
