Question 1.1 - Where are the companies with the best infrastructure to deal with mental health issues?

```python
companies=data[['state','benefits','care_options','wellness_program','seek_help']].reset_index(drop=True)
```

```python
companies
```

```python
for col in companies.columns:
    print(data[col].value_counts())
```

```python
di={'Yes':1.0,'No':0.0,'Don\'t know':-0.75,'Not sure':-0.75}
companies.replace({'benefits': di,'care_options':di,'wellness_program':di,'seek_help':di},inplace=True)
```

```python
companies
```

```python
companies['valuation']=3*companies['benefits']+2*(companies['seek_help']+companies['care_options'])+companies['wellness_program']
companies
```

```python
where=companies[['state','valuation']]
where=where.groupby('state').sum().reset_index()
where.sort_values(by='state', ascending=False,inplace=True,ignore_index=True)
where.rename(columns={'state':'loc'},inplace=True)
where
```

```python
benefits=companies.groupby('state').sum().reset_index()
benefits
```

```python
table=pd.concat([rate, where], axis=1)
table
```

```python
table.drop(columns=['loc'],inplace=True)
```

```python
table.sort_values(by='rate',ascending=False,inplace=True,ignore_index=True)
```

```python
table.to_csv('linear_regression.csv')
```

```python
table=table[table['rate']>=20]
```

```python
labels = table['state']
y = table['rate']
z = table['valuation']

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(20,5))
rects1 = ax.bar(x - width/2, y, width, label='Number of patients',color='darkcyan')
rects2 = ax.bar(x + width/2, z, width, label='Resources',color='blue')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Scores by state')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.savefig("number_of_patients_vs_resources")

plt.show()
```
