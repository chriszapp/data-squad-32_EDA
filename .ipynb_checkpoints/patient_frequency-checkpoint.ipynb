```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('frequency.csv',index_col=[0])
```

```python
data
```

```python
data=data[['state','family_history','treatment']]
data=data.groupby(['state','family_history','treatment']).size().reset_index().rename(columns={0:'count'})
data.head(50)
```

```python
fam_treatment=data[(data['treatment']==1.0) & (data['family_history']==1.0)].reset_index(drop=True)
fam_treatment
```

```python
fig, ax = plt.subplots(figsize=(20,10))

# Example data
people = fam_treatment['state']
y_pos=range(0,35)

ax.barh(y_pos,fam_treatment['count'], align='center',color='green')
ax.set_yticks(y_pos)
ax.set_yticklabels(people)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Number of people')
ax.set_title('Number of people that did treatment and have family history')

for i, v in enumerate(fam_treatment['count']):
    ax.text(v+0.1, i + .25, str(v), color='red', fontweight='bold')

plt.show()
```
